# Consignes

Cet énoncé comporte deux problèmes et deux exercices. Les deux problèmes qu'il comporte sont plutôt ouverts et relativement difficiles.  Les deux exercices sont des classiques.
Vous avez, comme déjà dit en cours, droit à toute la matière que vous souhaitez (IAs, notes personnelles, corrigés des TPs).

Avant de commencer, assurez-vous de travailler sur votre version synchronisée du dépôt ML_S5_etudiants et pensez à sauvegarder régulièrement pendant l'épreuve.

La durée est de trois heures exactement. Prévoyez 5 minutes à la fin pour envoyer à l'adresse tpdeeplearning@gmail.com une archive .tar (ou .zip) contenant un notebook par problème et par exercice (soit quatre notebooks au maximum).

Vous serez évalués sur les points suivants:

- la rédaction. Si vous utilisez l'IA, fournissez les prompts qui vous ont amené au résultat, et mettez en avant votre esprit critique (vous pouvez éventuellement citer la réponse fournie par la machine). Dans tous les cas, présentez votre démarche. C'est la qualité de votre production qui sera jugée avant tout. Il n'est donc pas primordial d'avoir tout réalisé.

- la propreté du notebook : je dois pouvoir comprendre assez vite comment vous avez abordés les questions, pouvoir faire tourner le code sans problème et en comprendre les éléments.

Bon courage à tous !
